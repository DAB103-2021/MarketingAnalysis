```python
import pandas as pd
import os
import tweepy
from dotenv import load_dotenv

load_dotenv()
consumer_key = os.environ["API_KEY"]
consumer_secret = os.environ["API_KEY_SECRET"]
access_token = os.environ["ACCESS_TOKEN"]
access_token_secret = os.environ["ACCESS_TOKEN_SECRET"]

auth = tweepy.OAuth1UserHandler(
  consumer_key, 
  consumer_secret, 
  access_token, 
  access_token_secret
)

api = tweepy.API(auth)
```

```python
extracted_pages = []

for page in tweepy.Cursor(api.search_tweets, 
                            "cybertruck", 
                            lang="en",tweet_mode='extended',
                            count=200000).pages(200000):
    extracted_pages.append(page)
    
    
extracted_tweets_from_pages = []

for page in extracted_pages:
    extracted_tweets_from_pages += page
pd.set_option("display.max_colwidth", None)   
db = pd.DataFrame(columns=['username','location','retweetcount','text','hashtags','created_at'])    
i = 1 
for tweet in extracted_tweets_from_pages:
    username = tweet.user.screen_name
    location = tweet.user.location
    retweetcount = tweet.retweet_count
    hashtags = tweet.entities['hashtags']
    created_at = tweet.created_at
    try:
        text = tweet.retweeted_status.full_text
    except AttributeError:
        text = tweet.full_text
    ith_tweet = [username,location,retweetcount,text,hashtags,created_at]
    db.loc[len(db)] = ith_tweet
    i = i+1
db    
```

```python
db.to_csv('tesla_cybertruck_tweets.csv')
```
